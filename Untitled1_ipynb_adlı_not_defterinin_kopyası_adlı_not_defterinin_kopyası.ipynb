*** Select several regression models and train them with the preprocessed data.Examine the performances of the selected models using cross validation.
Choose the best performing model***

```python
from sklearn.linear_model import LinearRegression, Lasso, BayesianRidge
models = [LinearRegression(), Lasso(), BayesianRidge()]
for model in models:
    model.fit(X_train, y_train)
```

```python
for model in models:
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    print(f'Model: {model.__class__.__name__}, MSE: {mse}, R2: {r2}')
```

```python
for model in models:
    scores = cross_val_score(model, X_train_scaled, y_train, cv=5, scoring='neg_mean_squared_error')
    rmse_scores = np.sqrt(-scores)

    print(f"Model: {model.__class__.__name__}")
    print(f"RMSE Scores: {rmse_scores}")
    print(f"Average RMSE: {rmse_scores.mean():.2f}")
    print("="*30)
```

**Choose the best performing model**

```python
best_model = RandomForestRegressor(random_state=42)
```

```python
best_model = models[np.argmin([score.mean() for score in rmse_scores])]
print(f"Best Performing Model: {best_model.__class__.__name__}")
```
